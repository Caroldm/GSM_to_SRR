From the GSM IDs we first get metadata files from each sample. Then find the SRA links from each metadata file, crawl 
the linked page and output the SRR for each GSM. 



Will read over all the files in the folder. And look the defined lines.

```python
import os
import re

path = '/Users/Carol/Desktop/gsm_samples/'


for filename in os.listdir(path):
    with open(path + filename,'rU') as data, open('gsm_to_urls.txt','w') as output:
        data=data.readlines()
        words = ['/sra-instant/reads/']
        keep = re.compile(r"|".join(words))
        #print(filename)
        for line in data:
            if keep.search(line):
                l = line.replace('!Sample_supplementary_file_',' ') 
                fname = filename.strip('.soft')
                print(fname + l)
                output.write(fname + l + '\n')
                
```

```python
import csv

with open('gsm_ftp.csv','rU') as f1, open('gsm_srr.csv','w')as out:
    data = csv.reader(f1)
    for line in data:
        #print(line)
        GSM=line[0]
        FTP=line[3]
        print (GSM + ' , '+ FTP)
        out.write(GSM + ' , ' + FTP +'\n')
```

```python
import re
import csv
import urllib


urls=[]
data = csv.reader(open('gsm_srr.csv','rU'), delimiter=',')    
for line in data:
    urls.append(line[1])
    for url in urls:
        htmlfile =  urllib.urlopen(url)
        htmltext =  htmlfile.readlines() 
        for line in htmltext:
            print(url + ' , ' + line)
        
```
